## Schools

```python
df = gpd.read_file('./geospatial/Non_Public_Schools.geojson')
dfp = gpd.read_file('./geospatial/Public_Schools.geojson')
df.shape
df.iloc[:, :-1].nunique()
# df['ScopeOfWor'].unique()

fig, ax = plt.subplots(1, figsize = (6, 8))
df.plot(marker = '.', column = 'TYPE', legend = True, ax = ax)
dfp.plot(marker = '*', column = 'SCH_TYPE', legend = True, ax = ax)
```
